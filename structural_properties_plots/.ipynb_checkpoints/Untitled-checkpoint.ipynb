# LUZ7

```python
pos_t5_curvature = pd.read_csv('LUZ7_t5_curvature.plus.csv', index_col=0)[50:-90]
pos_t10_curvature = pd.read_csv('LUZ7_t10_curvature.plus.csv', index_col=0)[50:-90]
pos_t20_curvature = pd.read_csv('LUZ7_t20_curvature.plus.csv', index_col=0)[50:-90]

pos_t5_stability = pd.read_csv('LUZ7_t5_stability.plus.csv', index_col=0)[50:-90]
pos_t10_stability = pd.read_csv('LUZ7_t10_stability.plus.csv', index_col=0)[50:-90]
pos_t20_stability = pd.read_csv('LUZ7_t20_stability.plus.csv', index_col=0)[50:-90]

pos_t5_bendability = pd.read_csv('LUZ7_t5_bendability.plus.csv', index_col=0)[50:-90]
pos_t10_bendability = pd.read_csv('LUZ7_t10_bendability.plus.csv', index_col=0)[50:-90]
pos_t20_bendability = pd.read_csv('LUZ7_t20_bendability.plus.csv', index_col=0)[50:-90]

randomSeq_curvature = pd.read_csv('random_seq_LUZ7_t5_curvature.plus.csv', index_col=0)[50:-90]
randomSeq_stability = pd.read_csv('random_seq_LUZ7_t5_stability.plus.csv', index_col=0)[50:-90]
randomSeq_bendability = pd.read_csv('random_seq_LUZ7_t5_bendability.plus.csv', index_col=0)[50:-90]
```

# comparison of structural properties between timesteps

```python
ax = pos_t5_curvature.loc[:, pos_t5_curvature.columns!='mean'].plot(use_index=True, color="green")
ax = pos_t10_curvature.loc[:, pos_t10_curvature.columns!='mean'].plot(use_index=True, color="red", ax=ax)
pos_t20_curvature.loc[:, pos_t20_curvature.columns!='mean'].plot(use_index=True, color="blue", ax=ax)
ax.legend('')
plt.title("Curvature")
```

```python
ax = pos_t5_curvature.loc[:, pos_t5_curvature.columns=='mean'].plot(use_index=True, color="green")
ax = pos_t10_curvature.loc[:, pos_t10_curvature.columns=='mean'].plot(use_index=True, color="red", ax=ax)
ax = pos_t20_curvature.loc[:, pos_t20_curvature.columns=='mean'].plot(use_index=True, color="blue",label="late", ax=ax)
randomSeq_curvature.loc[:, randomSeq_curvature.columns=='mean'].plot(use_index=True, color="black",label="randomSeq")
ax.legend(["early","middle","late"])
plt.title("Curvature")
```

```python
ax = pos_t5_stability.loc[:, pos_t5_stability.columns=='mean'].plot(use_index=True, color="green")
ax = pos_t10_stability.loc[:, pos_t10_stability.columns=='mean'].plot(use_index=True, color="red", ax=ax)
pos_t20_stability.loc[:, pos_t20_stability.columns=='mean'].plot(use_index=True, color="blue", ax=ax)
ax.legend(["early","middle","late"])
plt.title("Stability")
```

```python
ax = pos_t5_bendability.loc[:, pos_t5_bendability.columns!='mean'].plot(use_index=True, color="green")
ax = pos_t10_bendability.loc[:, pos_t10_bendability.columns!='mean'].plot(use_index=True, color="red", ax=ax)
pos_t20_bendability.loc[:, pos_t20_bendability.columns!='mean'].plot(use_index=True, color="blue", ax=ax)
ax.legend('')
plt.title("Bendability")
```
