```python
# dependencies
from dnacurve import CurvedDNA
from Bio import SeqIO
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fasta_sequences = SeqIO.parse(open(snakemake.input[0]),'fasta')
print(snakemake.config["TSS sequence extraction"]["upstream"])
df = pd.DataFrame(index=range(snakemake.config["TSS sequence extraction"]["upstream"],-1*(snakemake.config["TSS sequence extraction"]["downstream"]), -1))
for fasta in fasta_sequences:
    nameSeq, sequence = fasta.id, str(fasta.seq)
    result = CurvedDNA(sequence, 'trifonov', name=nameSeq)
    df[str(result.name)] = result.curvature[0,:]
df.plot(use_index=True)
plt.xlim(snakemake.config["TSS sequence extraction"]["upstream"] - 10, -1*(snakemake.config["TSS sequence extraction"]["downstream"])+11)
plt.savefig(snakemake.output[0])
```
